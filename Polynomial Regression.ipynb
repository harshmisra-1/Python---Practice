```python
import matplotlib.pyplot as plt
from scipy import stats

slope, intercept, r_value, p_value, std_err = stats.linregress(df['CombMPG'],df['HwyMPG'])

def predict(x):
    return slope * x + intercept

fitLine = predict(df['CombMPG'])

plt.scatter(df['CombMPG'],df['HwyMPG'])
plt.plot(df['CombMPG'], fitLine, c='r')
plt.show()
```

```python
x = np.array(df['CombMPG'])
y = np.array(df['HwyMPG'])

p4 = np.poly1d(np.polyfit(x, y, 4))
```

```python
import matplotlib.pyplot as plt

xp = np.linspace(10, 60, 100)
plt.scatter(x, y)
plt.plot(xp, p4(xp), c='r')
plt.show()
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y, p4(x))
r1 = r2_score(y,p1(x))
r10 = r2_score(y,p10(x))

print(r2)
```

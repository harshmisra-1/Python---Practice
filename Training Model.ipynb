```python
%matplotlib inline
import numpy as np
from pylab import *

np.random.seed(2)

pageSpeeds = np.random.normal(3.0, 1.0, 100)
purchaseAmount = np.random.normal(50.0, 30.0, 100) / pageSpeeds


scatter(pageSpeeds, purchaseAmount)
```

```python
trainX = pageSpeeds[:80]
testX = pageSpeeds[80:]

trainY = purchaseAmount[:80]
testY = purchaseAmount[80:]
```

```python
scatter(trainX, trainY)
```

```python
scatter(testX, testY)
```

```python
x = np.array(trainX)
y = np.array(trainY)

p4 = np.poly1d(np.polyfit(x, y, 4))
```

```python
import matplotlib.pyplot as plt

xp = np.linspace(0, 7, 100)
axes = plt.axes()
axes.set_xlim([0,7])
axes.set_ylim([0, 200])
plt.scatter(x, y)
plt.plot(xp, p4(xp), c='r')
plt.show()
```

```python
testx = np.array(testX)
testy = np.array(testY)

axes = plt.axes()
axes.set_xlim([0,7])
axes.set_ylim([0, 200])
plt.scatter(testx, testy)
plt.plot(xp, p4(xp), c='r')
plt.show()
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(testy, p4(testx))

print(r2)
```

Not bad but then the datset was way to small. 

```python
from sklearn.metrics import r2_score

r2 = r2_score(np.array(trainY), p4(np.array(trainX)))

print(r2)
```

Performs well on the training dataset because the model was trained on that only. 

We tried on different polynomial values, we see that if we fit a linear model with degree = 1, the r squared values
are better for train but not for training. If we increase the degree to 4 it fits the training data very well but fails
to give good accuracy for testing dataset.
